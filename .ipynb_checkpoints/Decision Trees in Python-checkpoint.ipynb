```python
# !pip install DecisionTreeClassifier
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
import numpy as np
```

# Importing Data

```python
headers = ['age', 'sex','cp','restbp','chol','fbs','restecg','thalach','exang','oldpeak', 'slope', 'ca', 'thal','hd']
```

```python
data = pd.read_csv("./processed.cleveland.data", header = None)
data.columns=headers
```

```python
data.describe()
```

```python
data.dtypes
```

```python
data['ca'].unique()
```

```python
data['thal'].unique()
```

```python
filtered_data = data.loc[~((data['ca']=='?') | (data['thal']=='?'))]
```

# Splitting data

```python
X = filtered_data.drop(columns = ['hd'])
y = filtered_data['hd'].copy()
```

```python
X = pd.get_dummies(X, columns=['cp', 'thal', 'restecg', 'slope'])
```

```python
y.loc[y>0] = 1
y.unique()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

# Classifier

```python
dct = DecisionTreeClassifier(random_state=42) 
```

```python
tree = dct.fit(X_train, y_train)
```

```python
y_preds = tree.predict(X_test)
print(accuracy_score(y_preds, y_test))
```

```python
cm = confusion_matrix(y_preds,y_test)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_preds, y_test))
```

## Finding Best Pruning Alpha

```python
path = dct.cost_complexity_pruning_path(X_train, y_train)
ccp_alphas, impurities = path.ccp_alphas, path.impurities
ccp_alphas = ccp_alphas[:-1]

clf_dts = []
for ccp_alpha in ccp_alphas:
    clf_dt = DecisionTreeClassifier(random_state=0, ccp_alpha=ccp_alpha)
    clf_dt.fit(X_train, y_train)
    clf_dts.append(clf_dt)
```

```python
train_scores = [clf_dt.score(X_train, y_train) for clf_dt in clf_dts]
test_scores = [clf_dt.score(X_test, y_test) for clf_dt in clf_dts]

fig, ax = plt.subplots()
ax.set_xlabel("alpha")
ax.set_ylabel("accuracy")
ax.set_title("Accuracy vs alpha for training and testing sets")
ax.plot(ccp_alphas, train_scores, marker='o', label="train", drawstyle="steps-post")
ax.plot(ccp_alphas, test_scores, marker='o', label="test", drawstyle="steps-post")
ax.legend()
plt.show()
```

```python
clf_dt = DecisionTreeClassifier(random_state=42, ccp_alpha=0.016)
scores = cross_val_score(clf_dt, X_train, y_train, cv=5)
df = pd.DataFrame(data={'tree': range(5), 'accuracy': scores})

df.plot(x='tree', y='accuracy', marker='o', linestyle='--')
```

```python
alpha_loop_values = []
for ccp_alpha in ccp_alphas:
    clf_dt = DecisionTreeClassifier(random_state=0, ccp_alpha=ccp_alpha)
    scores = cross_val_score(clf_dt, X_train, y_train, cv=5)
    alpha_loop_values.append([ccp_alpha, np.mean(scores), np.std(scores)])
    
alpha_results = pd.DataFrame(alpha_loop_values, 
                             columns=['alpha', 'mean_accuracy', 'std'])

alpha_results.plot(x='alpha', 
                   y='mean_accuracy', 
                   yerr='std', 
                   marker='o', 
                   linestyle='--')
```

```python
new_ccp_alphas = ccp_alphas[(ccp_alphas>0.014) & (ccp_alphas<0.015)]
new_ccp_alphas = float(new_ccp_alphas)
```

# Result with Best Alpha

```python
clf_dt_pruned = DecisionTreeClassifier(random_state=42, 
                                       ccp_alpha=new_ccp_alphas)
clf_dt_pruned = clf_dt_pruned.fit(X_train, y_train) 
```

```python
y_preds = clf_dt_pruned.predict(X_test)
print(accuracy_score(y_preds, y_test))
```

```python
cm = confusion_matrix(y_preds,y_test)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_preds, y_test))
```
